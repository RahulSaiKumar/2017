```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline 
# Helps make using plt.show() unnecessary.
```

# Introduction to the Dataset

```python
recent_grads = pd.read_csv("F:/Learning/recent_grads.csv")
rd_first_row = recent_grads.iloc[0]
print(recent_grads.head(5)) #Displays First five rows of dataset.
print(recent_grads.tail(5)) # Displays last five rows of dataset.
recent_grads.describe() #Displays the count, mean, std, min, max and quartiles of the table data.
```

# Cleaning the Dataset

```python
# Cleaning the dataset from Null values and having equal number of rows(data) for all columns.
raw_data_count = recent_grads.count()
recent_grads = recent_grads.dropna()
cleaned_data_count = recent_grads.count()
```

# Exploring Correlations Between Columns

```python
recent_grads.plot(x = 'Sample_size', y = 'Employed', kind = 'scatter', title = 'Employed vs. Sample_size', figsize = (5,10))
```

```python
ax2 = recent_grads.plot(x = 'Sample_size', y = 'Unemployment_rate', kind = 'scatter')
ax2.set_title('Sample_size vs. Unemployment_rate')
```

```python
ax4 = recent_grads.plot(x = 'ShareWomen', y = 'Unemployment_rate', kind = 'scatter')
ax4.set_title('ShareWomen vs. Unemployment_rate')
```

```python
ax5 = recent_grads.plot(x = 'Men', y = 'Median', kind = 'scatter')
ax5.set_title('Men vs. Median')
```

```python
ax6 = recent_grads.plot(x = 'Women', y = 'Median', kind = 'scatter')
ax6.set_title('Women vs. Median')
```

## Findings

From the above scatter plots, there seems to be a strong positive correlation between the sample size of students considered and employed rate. 

There seems to be a strong positive correlation between smaller sample sizes and Unemployment rate, but the correlation becomes almost negligible considering bigger sample sizes. 

There is absolutely no correlation between the Share of Women Students in a Major and their Unemployment_rate.

From the plot between men vs. median and women vs. median, we notice that overall, students from majors with low student strength tend to earn a higher median salary than students from majors with larger student strength.

# Do students that majored in subjects that were majority female make more money?

```python
majorly_women = recent_grads[recent_grads['Women']>recent_grads['Men']] 
# copying all records of dataset where women count is more than male count and then sorting records based on Share of Women
majorly_women = majorly_women.sort_values('ShareWomen', ascending = False)
majorly_women['Major_category'].value_counts()
```

```python
women_major_category = ['Education','Biology & Life Science','Health','Humanities & Liberal Arts','Psychology & Social Work']
for each in women_major_category:
    print(each)
    print(majorly_women[majorly_women['Major_category'] == each]['Median'].mean()); print('\n')
```

```python
minor_women = recent_grads[recent_grads['ShareWomen']<.5]
#Copying all records where women are the minority and then sorting based on Share of Women
minor_women = minor_women.sort_values('ShareWomen')
minor_women['Major_category'].value_counts()
```

```python
minor_women[minor_women['Major_category'] == 'Engineering']['Median'].mean()
```

```python
ax8 = majorly_women.plot(kind = 'scatter', x ='Women', y = 'Median', c = 'blue', label = 'Majority Women')
minor_women.plot(kind = 'scatter', x = 'Women', y = 'Median', c = 'red', label = 'Minority Women', ax=ax8)
ax8.set_title('Comparison of Median Salaries of Women in Majors with majority women and minority women')
```

## Findings

The plot shows that the median salary of women in women majority majors is less when compared to majors where women are in the minority. 

This seems right as women prefer streams like Education, Life Sciences, Health, Liberal Arts, Psychology which have a average median salary from 30000 USD to 37000 USD. Whereas streams like Engineering with a low share of women students pay a higher average median salary around 57000 USD.

## Compare the number of men with the number of women in each category of majors

```python
major_male =[] 
major_female=[]
for degree in major_cat: # Iterating to find the number of men and women in each major category and copying it to respective var
    major = recent_grads[recent_grads['Major_category'] == degree]
    major_male.append(major['Men'].sum())
    major_female.append(major['Women'].sum())
bar_positions = np.arange(len(major_cat))
tick_positions = bar_positions+.125
fig,ax11 = plt.subplots(figsize = (8,8))
ax11.bar(bar_positions, major_male, .25, color = 'blue', label = 'Men')
ax11.bar(bar_positions+.25, major_female, .25, color = 'red', label = 'Women')
ax11.set_xticks(tick_positions)
ax11.set_xticklabels(major_cat, rotation = 90)
ax11.legend(loc = "upper right")
ax11.set_title("Women Strength vs. Men Strength in each Major Category")
```

## Findings

The plot shows that some major categories are preferred more by women and some others are preferred more by men. The Education category seems to be totally dominated by women. In categories like Business, Physical Sciences, Law, Industrial Arts, Social Sciences, Agriculture there seems to be an equal distribution of men and women students
