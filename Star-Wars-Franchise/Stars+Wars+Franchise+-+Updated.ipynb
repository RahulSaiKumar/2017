## Introduction to the Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
star_wars = pd.read_csv("F:/Learning/Guided Project- Star Wars Survey (1)/star_wars.csv", encoding ="ISO-8859-1")
```

```python
star_wars = star_wars[pd.notnull(star_wars['RespondentID'])]
```

```python
star_wars.head()
```

```python
star_wars.shape
```

```python
star_wars.columns
```

## Cleaning and Mapping Yes/No Columns

```python
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts()
```

```python
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts()
```

```python
boolean = {'Yes':True, 'No':False}
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(boolean)
```

```python
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(boolean)
```

```python
star_wars.head()
```

## Cleaning and Mapping Checkbox Columns

```python
seen_movie = {
    'Star Wars: Episode I  The Phantom Menace' : True,
    'Star Wars: Episode II  Attack of the Clones' : True,
    'Star Wars: Episode III  Revenge of the Sith' : True,
    'Star Wars: Episode IV  A New Hope' : True,
    'Star Wars: Episode V The Empire Strikes Back' : True,
    'Star Wars: Episode VI Return of the Jedi' : True,
    np.nan : False
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(seen_movie)
star_wars.head()
```

```python
star_wars = star_wars.rename(columns = {
    "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
    "Unnamed: 4": "seen_2",
    "Unnamed: 5": "seen_3",
    "Unnamed: 6": "seen_4",
    "Unnamed: 7": "seen_5",
    "Unnamed: 8": "seen_6"
     })
star_wars.head()
```

## Cleaning the Ranking Columns

```python
star_wars = star_wars.rename(columns = {
    'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.' : 'ranking_1',
    'Unnamed: 10' : 'ranking_2',
    'Unnamed: 11' : 'ranking_3',
    'Unnamed: 12' : 'ranking_4',
    'Unnamed: 13' : 'ranking_5',
    'Unnamed: 14' : 'ranking_6'
})

star_wars.head()
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

## Ranking

```python
seen_all_6 = star_wars[(star_wars['seen_1'] == True) & (star_wars['seen_2'] == True) & (star_wars['seen_3'] == True) & (star_wars['seen_4'] == True) & (star_wars['seen_5'] == True) & (star_wars['seen_6'] == True)]
seen_all_6.head()
```

```python
num_cols = ['Return of the Jedi','The Empire Strikes Back','A New Hope','Revenge of the Sith','Attack of the Clones','The Phantom Menace']
cols = [6.0,5.0,4.0,3.0,2.0,1.0]
seen_all_6_best_per = (seen_all_6['ranking_1'].value_counts()/sum(seen_all_6['ranking_1'].value_counts()))*100
bar_widths = seen_all_6_best_per[cols].values
bar_pos = np.arange(6) + 1
tick_pos = range(1,7)
fig = plt.figure( figsize = (6,6))

ax1 = fig.add_subplot(1,1,1)
ax1.barh(bar_pos,bar_widths,.5)
ax1.set_yticks(tick_pos)
ax1.set_yticklabels(num_cols,fontsize = 15)
ax1.set_title('The Best Star Wars Movie According to Survey Respondents?',fontsize = 15)
ax1.set_xlabel('Percentage',fontsize = 15)
ax1.tick_params(left = 'off',labelsize = 15)
ax1.spines["right"].set_visible(False)    
ax1.spines["left"].set_visible(False)
ax1.spines["top"].set_visible(False)    
ax1.spines["bottom"].set_visible(False)
```

## Findings

Suprisingly this set of sample responded have rated the 6th movie "Return of the Jedi" as the highest as opposed to the global consensus that the 5th film "The Empire Strikes Back" is the best. We shall try to get a better understanding by dividing the ratings given into top, mid and last third for more accurate insights.

One other thing noticable is that the Fans seem to be divided into two groups, While one group prefers the older star wars movies, another prefers the newer trilogy. This can be inferred from the data, where the major share of respondents who have rated "Return of the Jedi" first, have also given the highest rating for the 4th and 5th movies. While the second group have mostly rated the newer three as first.

```python
top_third = seen_all_6['ranking_1']
top_third = top_third.append(seen_all_6['ranking_2'])

mid_third = seen_all_6['ranking_3']
mid_third = mid_third.append(seen_all_6['ranking_4'])

last_third = seen_all_6['ranking_5']
last_third = last_third.append(seen_all_6['ranking_6'])

top_third_per = (top_third.value_counts()/sum(top_third.value_counts())) *200
mid_third_per = (mid_third.value_counts()/sum(mid_third.value_counts())) *200
last_third_per = (last_third.value_counts()/sum(last_third.value_counts())) *200
```

```python
print("\t\tHow Often Each Film was Rated in the Top, Mid and Last Third by the Respondents Who Watched All 6 Movies")

num_cols = ['Return of the Jedi','The Empire Strikes Back','A New Hope','Revenge of the Sith','Attack of the Clones','The Phantom Menace']
cols = [6.0,5.0,4.0,3.0,2.0,1.0]
bar_widths1 = top_third_per[cols].values
bar_widths2 = mid_third_per[cols].values
bar_widths3 = last_third_per[cols].values
bar_pos = np.arange(6) + 1
tick_pos = range(1,7)
fig = plt.figure( figsize = (18,6))

ax1 = fig.add_subplot(1,3,1)
ax1.barh(bar_pos,bar_widths1,.5)
ax1.set_yticks(tick_pos)
ax1.set_yticklabels(num_cols,fontsize = 15)
ax1.set_title('Ranked in the Top Third',fontsize = 15)
ax1.set_xlabel('Percentage',fontsize = 15)
ax1.tick_params(left = 'off',labelsize = 15)
ax1.spines["right"].set_visible(False)    
ax1.spines["left"].set_visible(False)
ax1.spines["top"].set_visible(False)    
ax1.spines["bottom"].set_visible(False)

ax2 = fig.add_subplot(1,3,2)
ax2.barh(bar_pos,bar_widths2,.5)
ax2.set_yticks([])
ax2.set_title('Ranked in the Middle Third',fontsize = 15)
ax2.tick_params(left = 'off',labelsize = 15)
ax2.set_xlabel('Percentage',fontsize = 15)
ax2.spines["right"].set_visible(False)    
ax2.spines["left"].set_visible(False)
ax2.spines["top"].set_visible(False)    
ax2.spines["bottom"].set_visible(False)

ax3 = fig.add_subplot(1,3,3)
ax3.barh(bar_pos,bar_widths3,.5)
ax3.set_yticks([])
ax3.tick_params(left = 'off',labelsize = 15)
ax3.set_title('Ranked in the Last Third',fontsize = 15)
ax3.set_xlabel('Percentage',fontsize = 15)
ax3.spines["right"].set_visible(False)    
ax3.spines["left"].set_visible(False)
ax3.spines["top"].set_visible(False)    
ax3.spines["bottom"].set_visible(False)

ax1.set_xlim(0,60)
ax2.set_xlim(0,60)
ax3.set_xlim(0,60)


plt.show()
```

## Findings

The plot shows us that while not many had rated the 5th movie as first, it was still rated in the top two by the most number of respondents who watched all 6 films. This is in aggreement to the global consensus that the 5th film in the series is clearly the best.

## Movie Ranking and View Count based on Star Trek Fandom

```python
star_trek_fans = star_wars[star_wars['Do you consider yourself to be a fan of the Star Trek franchise?'] == 'Yes']
star_trek_nonfans = star_wars[star_wars['Do you consider yourself to be a fan of the Star Trek franchise?'] == 'No']
```

```python
star_trek_fans[star_trek_fans.columns[9:15]].mean().plot.barh()
plt.title("Ranking of Star Wars Franchise by Star Trek Fans")
plt.show()

star_trek_nonfans[star_trek_nonfans.columns[9:15]].mean().plot.barh()
plt.title("Ranking of Star Wars Franchise by Star Trek Non-Fans")
plt.show()
```

Based on the plot both the fans and non-fans of the Star Trek franchise agree that the original Star Wars movies were better than the newer ones. On a closer look at the rankings given to the newer versions, Most Star Trek fans for the rankings 1,2 and seem to have ranked it with the original Star Wars movies(4, 5, 6), while a portion of the non-fans have rated the newer movies also high which brings the Ranking_1 to Ranking_3 columns to have a lower mean(3.5 to 4.3).

Each Ranking column is populated with the Star Wars Movie Order number. Since we are plotting using the mean, a higher mean implies more number of entries with (4,5,6) original movies.

```python
star_trek_fans[star_trek_fans.columns[3:9]].sum().plot.barh()
plt.title('View Count of Star Trek Fans Who Have Watched Star Wars')
plt.show()

star_trek_nonfans[star_trek_nonfans.columns[3:9]].sum().plot.barh()
plt.title('View Count of Star Trek Non-Fans Who Have Watched Star Wars')
plt.show()
```

There are a lot more Star Trek fans who have watched the Star Wars movies than non-fans. A possible reason for the big difference in view count of fans and non-fans could be that Star Trek non-fans do not prefer the Sci-fi Genre.
